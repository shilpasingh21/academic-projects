```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("Business_Review_Tip_Data.csv")
data
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import nltk
from sklearn.naive_bayes import MultinomialNB, BernoulliNB, GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn import preprocessing

# silencing the warnings
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

from nltk.corpus import stopwords
stopWords = set(stopwords.words('english'))
```

```python
textData = TfidfVectorizer(lowercase=True, strip_accents='unicode', stop_words=stopWords)
Xtext = textData.fit_transform(data['text'])
textCols = [('body_' + feature) for feature in textData.get_feature_names()]
```

```python
Xtext = pd.DataFrame(data=Xtext.todense(), columns=textCols)
y = data['class']
```

```python
# Taking the top 500 best words
selector = SelectKBest(chi2, k=500).fit(Xtext, y)
XtextTop500 = Xtext.iloc[:,selector.get_support(indices=True)]
XtextTop500
```

```python
XtipText = textData.fit_transform(data['tipText'])
tipTextCols = [('tip_' + feature) for feature in textData.get_feature_names()]
XtipText = pd.DataFrame(data=XtipText.todense(), columns=tipTextCols)
```

```python

# Taking the top 500 best words
selector = SelectKBest(chi2, k=500).fit(XtipText, y)
XtiptextTop500 = XtipText.iloc[:,selector.get_support(indices=True)]
XtiptextTop500
```

```python
data.columns
```

```python
newData = data.drop(['Unnamed: 0', 'text','tipText'], axis=1)
```

```python
# Join initial DataFrame Xraw and features extracted from body and title
X = pd.concat([newData, XtextTop500, XtiptextTop500], axis=1)
X
```

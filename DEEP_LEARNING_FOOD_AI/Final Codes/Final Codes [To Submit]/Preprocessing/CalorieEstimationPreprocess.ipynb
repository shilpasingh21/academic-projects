```
pickletraindata_Y2 = 'gdrive/My Drive/DL Project/PickleFiles/pickleTestFull_Y2.pkl'
calorie_file = 'gdrive/My Drive/DL Project/Data/SplitAndIngreLabel/IngredientCalorie.txt'

train_sum_calorie_file = 'gdrive/My Drive/DL Project/PickleFiles/pickleTest_Y3_sum.pkl' # Y3 is the sum of all the calorie for ingredients for a food sample
train_avg_calorie_file = 'gdrive/My Drive/DL Project/PickleFiles/pickleTest_Y3_avg.pkl' # Y3 is the average of all the calorie for ingredients for a food sample

```

```
# Load the Y2 data in session

train_Y2_list = pickle.load(open(pickletraindata_Y2, 'rb' ))
train_Y2 = np.array(train_Y2_list)

for index in range(train_Y2.shape[0]):
  templist = train_Y2[index]
  templist_0 = [0 if x==-1 else x for x in templist]
  train_Y2[index] = templist_0

print(train_Y2.shape)
```

```
# Load the calorie reference file for ingredients in a list

calorielist = []
with open(calorie_file) as file:
  for line in file:
    calorielist.append(line.strip())
```

```
# Preapre Y3 as sum of the calorie for 1 unit

caloriesum = []
for i in range(len(train_Y2)):
  sum =0
  for j in range(len(train_Y2[i])):
    if(train_Y2[i][j] == 1):
     sum = sum + int(calorielist[j])
  caloriesum.append(sum)
```

```
print(len(caloriesum))
```

```
with open(train_sum_calorie_file, 'wb') as file:
    pickle.dump(caloriesum, file)
```

```
# Preapre Y3 as average of the calorie for 1 unit

calorieavg = []
for i in range(len(train_Y2)):
  sum =0
  N =0
  for j in range(len(train_Y2[i])):
    if(train_Y2[i][j] == 1):
      N = N + 1
      sum = sum + int(calorielist[j])
  avg = sum/N
  calorieavg.append(avg)
```

```
print(len(calorieavg))
```

```
with open(train_avg_calorie_file, 'wb') as file:
    pickle.dump(calorieavg, file)
```
